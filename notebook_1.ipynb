```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
trader = pd.read_csv("historical_data.csv")
sentiment = pd.read_csv("fear_greed_index.csv")
```

```python
trader.head()
```

```python
sentiment.head()
```

```python
trader.columns = trader.columns.str.strip().str.lower()
sentiment.columns = sentiment.columns.str.strip().str.lower()

```

```python
trader.head()
```

```python
sentiment.head()
```

```python
trader.drop(columns=['timestamp'], inplace=True)
sentiment.drop(columns=['timestamp'], inplace=True)



```

```python
trader.columns = trader.columns.str.replace(' ', '_')
```

```python
trader['timestamp_ist'] = pd.to_datetime(
    trader['timestamp_ist'],
    dayfirst=True,
    errors='coerce'
)
trader['date'] = trader['timestamp_ist'].dt.date

sentiment['date'] = pd.to_datetime(
    sentiment['date'],
    dayfirst=True,
    errors='coerce'
).dt.date




```

```python
sentiment.isnull().sum()
trader.isnull().sum()

```

```python
trader.drop_duplicates(inplace=True)
sentiment.drop_duplicates(inplace=True)
```

```python
trader.dtypes


```

```python
trader['trade_id'] = trader['trade_id'].astype(str)
```

```python
sentiment.dtypes
```

```python
Q1 = trader['execution_price'].quantile(0.25)
Q3 = trader['execution_price'].quantile(0.75)
IQR = Q3 - Q1

trader = trader[
    (trader['execution_price'] >= Q1 - 1.5 * IQR) &
    (trader['execution_price'] <= Q3 + 1.5 * IQR)
]
```

```python
final = pd.merge(trader, sentiment, on='date', how='inner')
```

```python
final.info()
print(final['classification'].value_counts())
final.head()
print(final.duplicated().sum())

```

```python
final.to_csv("cleaned_trader_sentiment.csv", index=False)
final.to_excel("cleaned_trader_sentiment.xlsx", index=False)
```
